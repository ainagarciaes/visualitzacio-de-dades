```python
# Import packages
import pandas as pd
import matplotlib.pyplot as plt

# import dataset
df = pd.read_csv("https://raw.githubusercontent.com/ainagarciaes/ainagarciaes.github.io/main/browsers.csv")

# set the figure size
plt.figure(figsize=(10, 10))

# plot the connected scatterplot
plt.plot(df.chrome, df.ie, '-', marker='o')

# add annotations in every 3 data points with a loop
for line in range(0, df.shape[0]):
     plt.annotate(
          df.year[line], 
          (df.chrome[line]+1, df.ie[line]+1),
          va='bottom',
          ha='center'
     )

# x axis label
plt.xlabel('google chrome')

# y axis label
plt.ylabel('internet explorer')

# show the graph
plt.show()
```
